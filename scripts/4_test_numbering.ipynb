# Chothia Numbering Tests of Generated Sequences

```python
import pandas as pd
from abnumber import Chain
```

```python
## Read in candidates data
candidates_df = pd.read_excel('../candidates.xlsx', sheet_name='Antibody Candidates')
```

```python
## Define numbering test
def test_numbering(seq:str="", chain_type:str="") -> str:
    try:
        chain = Chain(seq, scheme="chothia")
        if chain_type == "heavy":
            if not chain.is_heavy_chain():
                return False
            else:
                return True
        elif chain_type == "light":
            if not chain.is_light_chain():
                return False
            else:
                return True
        else:
            return True
    except Exception as e:
        print(f"Error with sequence {seq}: {e}")
        return False
```

```python
for idx, row in candidates_df.iterrows():
    antibody_id = row['antibody_id']
    h_chain = row['h_chain']
    l_chain = row['l_chain']

    ## Check H and L chain separately
    h_pass = test_numbering(seq = h_chain, chain_type = "heavy")
    l_pass = test_numbering(seq = l_chain, chain_type = "light")

    ## Check if both pass
    if (h_pass and l_pass):
        print(f"{antibody_id}: TRUE")
    else:
        print(f"{antibody_id}: FALSE")
```
