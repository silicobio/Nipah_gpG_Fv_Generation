# Parse Proteinbase Score (from API Response)

GET https://proteinbase.com/silicobio?pageSize=120&_rsc=va1tz

```python
import pandas as pd
import json
```

```python
json_path = '../data/candidates/competition_submissions/Proteinbase_scores_20251121.json'
```

```python
with open(json_path, 'r') as f:
    scores_dict = json.load(f)

# scores_dict
```

```python
scores_df = pd.json_normalize(scores_dict['proteins']['data'])

scores_df.head()
```

```python
eval_df = pd.json_normalize(scores_df['evaluations'])

eval_df['name'] = scores_df['name']

eval_df.head()
```

```python
eval_df_long = pd.melt(eval_df, 
                  id_vars=['name'], 
                  var_name='eval_id', 
                  value_name='eval_dict')

eval_df_long
```

```python
eval_df_long_pvt = pd.json_normalize(eval_df_long['eval_dict'])[['metric.slug', 'value']]

eval_df_long_pvt['name'] = eval_df_long['name']

## Filter to boltz2_ipsae metric
eval_df_long_pvt = eval_df_long_pvt[eval_df_long_pvt['metric.slug'].str.contains('boltz2_ipsae')]

eval_df_long_pvt.head()
```

```python
eval_df_pvt = eval_df_long_pvt.pivot(index='name', columns='metric.slug', values='value')
eval_df_pvt.tail()
```

```python
eval_df_pvt.to_csv('../data/candidates/competition_submissions/Proteinbase_scores_20251121_parsed.csv')
```
