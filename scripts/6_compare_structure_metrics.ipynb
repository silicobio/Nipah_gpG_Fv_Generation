# Compare Proteinbase and Locally-Derived Boltz-2 Structure Metrics

```python
import pandas as pd
import os, json
```

```python
## Read in candidates data
candidates_df = pd.read_excel('../candidates.xlsx', sheet_name='Antibody Candidates')
```

```python
scores_df = pd.DataFrame(
    columns=[
        'antibody_id',
        'confidence_score',
        'ptm',
        'iptm',
        'protein_iptm',
        'complex_plddt',
        'complex_iplddt',
        'complex_pde',
        'complex_ipde'
        ]
    )

for idx, row in candidates_df.iterrows():
    antibody_id = row['antibody_id']

    json_path = f"../data/candidates/structures_boltz2_frankenchain_comp/boltz_results_{antibody_id}/predictions/{antibody_id}/confidence_{antibody_id}_model_0.json"

    with open(json_path, 'r') as f:
        data = f.read()
        data_json = json.loads(data)

    score_df = pd.DataFrame(
        columns=[
            'antibody_id',
            'confidence_score',
            'ptm',
            'iptm',
            'protein_iptm',
            'complex_plddt',
            'complex_iplddt',
            'complex_pde',
            'complex_ipde'
            ],
        data=[[
                antibody_id,
                data_json['confidence_score'],
                data_json['ptm'],
                data_json['iptm'],
                data_json['protein_iptm'],
                data_json['complex_plddt'],
                data_json['complex_iplddt'],
                data_json['complex_pde'],
                data_json['complex_ipde']
                ]]
        )

    scores_df = pd.concat([scores_df, score_df], ignore_index=True)

scores_df.head()
```

```python
scores_df.to_csv('boltz2_structure_metrics.csv', index=False)
```
